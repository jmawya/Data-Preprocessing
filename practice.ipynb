```python
df=pd.read_csv(r"C:\Users\Jannatul Mawya\OneDrive\Pictures\Thesis\impermium_verification_labels.csv")
print("Dataset\n", df.columns.tolist())
print(df["Comment"])
print(df["Insult"])
```

```python
text_col=df['Comment']
label_col=df['Insult']
if text_col is None or label_col is None :
    raise ValueError("Could not detect text/level columns")

#print(f"Detected text column:\n {text_col}")    
print('Comment\n',text_col)
print('Insult\n',label_col)
```

```python
df.dropna(inplace=True) #Remove any rows in your DataFrame that have missing values (NaN)
df['Insult']=df['Insult'].astype(int) #inplace=True → modifies the DataFrame directly, instead of creating a new one.
print(df)

```

```python
def clean_text(text):
    text=re.sub(r'http\S+|<.*?>', ' ',str(text).lower())
    tokens=word_tokenize(text)
    stop_word=set(stopwords.words("english"))
    filtered_token=[]
    for word in tokens:
        if word.isalpha() and word not in stop_word:  #isalpha for removing punctuation
            filtered_token.append(word)
    pos_tags=nltk.pos_tag(filtered_token)
    lemma=[]
    for word,pos in pos_tags:
           base_word=lemmatizer.lemmatize(word,gt(pos))
           lemma.append(base_word)
    return ' '.join(lemma)
```

```python
df['clean_comment'] = df['Comment'].apply(clean_text)
print(df)  
#.apply(clean_text) will process each comment individually, one row at a time.
```

```python
#CountVectorizer is a tool from scikit-learn that converts text 
#into numbers that machine learning models can understand.
vectorizer= CountVectorizer(max_features=5000,ngram_range=(1,2))
x=vectorizer.fit_transform(df['clean_comment'])
y=df['Insult']
print("Comment:\n",x)
print("Insult:\n",y)
```

```python
x_train, x_test,y_train, y_test=train_test_split(
    x,y,test_size=0.2, stratify=y, random_state=42)
```

```python
# ===============================
# Models
# ===============================
lr = LogisticRegression(max_iter=1000, solver='liblinear')
rf = RandomForestClassifier(n_estimators=200, random_state=42)
nb = MultinomialNB()

lr.fit(x_train, y_train)
rf.fit(x_train, y_train)
nb.fit(x_train, y_train)

voting = VotingClassifier(
    estimators=[('lr', lr), ('rf', rf)],
    voting='soft'
)
voting.fit(x_train, y_train)

adaboost = AdaBoostClassifier(
    estimator=LogisticRegression(max_iter=1000, solver='liblinear'),
    n_estimators=50,
    random_state=42
)
adaboost.fit(x_train, y_train)
```

```python
for name,model in [
    ('Logistic Regression',lr),
    ('Random Forest',rf),
    ('Naive Bayes',nb),
    ('Voting Ensemble',voting),
    ('AdaBoost',adaboost)
]:
    pr=model.predict(x_test)
    print(name)
    print("Accuracy:",accuracy_score(y_test,pr))
    print(classification_report(y_test,pr))
    print("Confution Matrix:\n",confusion_matrix(y_test, pr))
```
