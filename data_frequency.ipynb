```python
std= df.iloc[2]["Text"]
print(std)
```

```python
tokens = word_tokenize(std.lower())

print(tokens)
```

```python
freq_dist = FreqDist(tokens) # it's crate a dictionary list 
print(freq_dist)
```

```python
for word, freq in freq_dist.items():
    print(word + str(freq))
```
